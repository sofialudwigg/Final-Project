```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv('/Users/quinnchryss-connell/Desktop/Senior Campaign/CountryData.csv')
X1 = dataset.iloc[:, 1:8].values
X2 = dataset.iloc[:, 9:].values
y = dataset.iloc[:, 8].values
X = np.concatenate((X1,X2),axis=1)

dataset
```

```python
X
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 11:])
X[:, 11:] = imputer.transform(X[:, 11:])
```

```python
pd.DataFrame(X[:,10:]).head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X[:, 11:] = sc.fit_transform(X[:, 11:])
```

```python
pd.DataFrame(X).isnull().sum()
```

```python
pd.DataFrame(y).isnull().sum()
```

```python
y = np.nan_to_num(y,nan=pd.DataFrame(y).mean())
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
regressor.score(X_test, y_test)
```
